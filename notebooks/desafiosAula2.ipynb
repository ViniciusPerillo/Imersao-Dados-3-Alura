## Desafio 1

```python
cod_compostos = dados['droga'].value_counts().index[:9]
cod_compostos
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(16,8))
sns.set_theme(font_scale=1.5)
ax = sns.countplot(x = 'droga', data= dados.query('droga in @cod_compostos'), order=dados['droga'].value_counts().index[:9])
ax.set_title('9 compostos mais usados', fontdict= {'fontsize': 20, 'fontweight': 'bold'})
label_fontdict = {'fontsize': 18}
ax.set_ylabel('Quantidade', fontdict=label_fontdict)
ax.set_xlabel('Compostos', fontdict=label_fontdict)
plt.show()
```

## Desafio 2 e 3


```python
label_fontdict = {'fontsize': 18}
plt.figure(figsize=(16,6))
sns.set_theme(font_scale=1.5)
ax = sns.histplot(x= 'g-0', data=dados, bins=1000, kde=True)
ax.set_title('Histograma dos dados em g0', fontdict= {'fontsize': 20, 'fontweight': 'bold'})
ax.set_ylabel('Quantidade', fontdict=label_fontdict)
ax.set_xlabel('Expressão gênica', fontdict=label_fontdict)
plt.show()
```

## Desafio 6

```python
plt.figure(figsize=(10,8))
sns.set_theme(font_scale=1.5)
ax = sns.boxenplot(x='g-0', y='tratamento', data=dados)
ax.set_title('Expressão do gene 0 de acordo com o tipo de tratamento', fontdict={'fontsize':20, 'fontweight':'bold'})
ax.set_xlabel('Expressão gênica')
ax.set_ylabel('Tratamento', fontdict=label_fontdict)
plt.show()
```
