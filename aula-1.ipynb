Durante a aula foi passado essa base de dados sobre algum experimento.

```python
import pandas as pd
dados_exp_url =  'dados/dados_experimentos.zip'
dados = pd.read_csv(dados_exp_url, compression = 'zip')
dados
```

Com uma breve instrução na aula a fonte dos dados da tabela foram contextualizados:

Testagem de farmacos em cultura de células. Dessa forma pode-se supor que cada linha seja um experimento.

## Analisando as colunas

- ###Id:
O id é um código único para cada experimento, tornando mais fácil a busca por um em especifico.
- ###Tratamento:
  


```python
dados['tratamento'].unique()
```

```python
dados['tratamento'].value_counts()
```

```python
import matplotlib.pyplot as plt
graphic_data = dados['tratamento'].value_counts(normalize=True)
colors = ['Orange','Blue']
labels = ['Com Droga', 'Com Controle']
explode = [0,0.2]

plt.pie(graphic_data,colors=colors,labels=labels,explode=explode, autopct= '%1.1f%%', shadow=True, textprops= {'fontsize': 12})
plt.title('Proporção dos dados na coluna tratamento', {'fontsize': 14, 'fontweight': 'bold'})
```

A coluna tratamento apresenta dois tipos de dados: com controle e com droga, eles apresentam um ocorrência bem diferente, isso é de se esperar já que eles representam os experimentos em que houve a testagem dos farmacos e os que foram usado como grupo controle*. 

*Grupo controle é um participante que participará dos mesmos preparativos porém não receberá a droga, assim podendo excluir dados provenientes da preparação e manipulação do experimento.

- ###Tempo: 

```python
dados['tempo'].unique()
```

```python
dados['tempo'].value_counts().plot.barh(title= 'Ocorrência dos dados da coluna tempo', fontsize=14, xlabel='tempo(horas)')
```

A coluna tempo apresenta três tipos de dados: 24, 48 e 72 que representam o tempo de obsevação do teste, diferente da coluna anterior, as proporções são mais próximas.

- ###Dose:


```python
dados['dose'].unique()
```

```python
dados['dose'].value_counts(ascending=True).plot.barh(title= 'Ocorrência dos dados na coluna dose', fontsize=14, xlabel='Doses')
```

Os dados nessa coluna apresentam de duas formas: D1 e D2, eles provavelmente são as duas doses testadas.
